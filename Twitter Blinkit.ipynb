```python
blinkit_keywords = [

    # üì¶ Service / Delivery Issues
    "Blinkit Complaint",
    "Blinkit late delivery",
    "Blinkit delivery issue",
    "Blinkit order not delivered",
    "Blinkit delivery delay",
    "Blinkit missed delivery",
    "Blinkit wrong item",
    "Blinkit delivery boy issue",
    "Blinkit cancelled my order",
    "Blinkit not delivering",
    "Blinkit didn't deliver",
    "Blinkit delivery scam",
    "Blinkit service complaint",

    # üí≥ Refund / Payment Issues
    "Blinkit refund not received",
    "Blinkit refund delay",
    "Blinkit charged twice",
    "Blinkit payment failed",
    "Blinkit payment issue",
    "Blinkit not refunding",
    "Blinkit money not returned",
    "Blinkit deducted money",

    # ü•¶ Product Quality Issues
    "Blinkit expired product",
    "Blinkit rotten vegetables",
    "Blinkit damaged item",
    "Blinkit sent stale food",
    "Blinkit poor packaging",
    "Blinkit quality issue",
    "Blinkit not fresh",

    # üëé Negative Sentiment / Experience
    "Blinkit worst experience",
    "Blinkit terrible service",
    "Blinkit poor customer service",
    "Blinkit very bad experience",
    "Blinkit never ordering again",
    "Blinkit worst app ever",
    "Blinkit horrible service",
    "Blinkit frustrating",
    "Blinkit disappointed",
    "Blinkit cheated",
    "Blinkit scam",
    "Blinkit fraud",

    # üßë‚Äçüíª Customer Support Issues
    "Blinkit no response",
    "Blinkit not responding",
    "Blinkit support not working",
    "Blinkit customer care",
    "Blinkit help",
    "Blinkit complaint",
    "Blinkit issue"
]
```

```python
import time
import random
from bs4 import BeautifulSoup

all_tweets = set()
tweet_list = []

for query in tqdm(blinkit_keywords):
    
    wr_query = ("%20").join(query.split())
    start_date = "2025-02-01"
    end_date = "2025-04-17"
    
    link = f"https://x.com/search?q={wr_query}%20until%3A{end_date}%20since%3A{start_date}&src=typed_query&f=top"
    driver.get(link)
    
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        # Scroll to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.uniform(4, 8))
    
        # Scrape visible tweets
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        posts = soup.find_all('div', attrs={'data-testid': 'cellInnerDiv'})
    
        for post in posts:
            try:
                article = post.find('article', attrs={"data-testid": "tweet"})
                if not article:
                    continue
    
                p1 = article.find('div', class_="css-175oi2r r-1iusvr4 r-16y2uox r-1777fci r-kzbkwu")
                if not p1:
                    continue
    
                p2 = p1.find_all('div', recursive=False)
                if len(p2) < 2:
                    continue
    
                try:
                    complainant_name = p2[0].find_all('a')[0].text.strip()
                except:
                    complainant_name = ""
    
                try:
                    complainant_id = p2[0].find_all('a')[1].text.strip()
                except:
                    complainant_id = ""
    
                try:
                    date_posted = p2[0].find_all('a')[2].find('time').get('datetime')
                except:
                    date_posted = ""
    
                flag = 0
                for iimg in p2:
                    if iimg.find('img'):
                        have_img_vid = 'Yes'
                        flag = 1
                        break
                    else:
                        have_img_vid = "No"
    
    
                if "Replying to @" in p2[1].text:
                    try:
                        written_post = p2[2].text.strip()
                    except:
                        written_post = ""
    
                    if flag == 1:
                        try:
                            stats = p2[4].find('div', class_="css-175oi2r r-1kbdv8c r-18u37iz r-1wtj0ep r-1ye8kvj r-1s2bzr4")\
                                        .find_all('div', class_="css-175oi2r r-18u37iz r-1h0z5md r-13awgt0")
                            comments = stats[0].text.strip() if len(stats) > 0 else "0"
                            repost = stats[1].text.strip() if len(stats) > 1 else "0"
                            likes = stats[2].text.strip() if len(stats) > 2 else "0"
                            views = stats[3].text.strip() if len(stats) > 3 else "0"
                        except:
                            comments = repost = likes = views = "0"
                    else:
                        try:
                            stats = p2[3].find('div', class_="css-175oi2r r-1kbdv8c r-18u37iz r-1wtj0ep r-1ye8kvj r-1s2bzr4")\
                                        .find_all('div', class_="css-175oi2r r-18u37iz r-1h0z5md r-13awgt0")
                            comments = stats[0].text.strip() if len(stats) > 0 else "0"
                            repost = stats[1].text.strip() if len(stats) > 1 else "0"
                            likes = stats[2].text.strip() if len(stats) > 2 else "0"
                            views = stats[3].text.strip() if len(stats) > 3 else "0"
                        except:
                            comments = repost = likes = views = "0"
    
                    
    
                else:
                    try:
                        written_post = p2[1].text.strip()
                    except:
                        written_post = ""
    
                    if flag == 1:
                        try:
                            stats = p2[3].find('div', class_="css-175oi2r r-1kbdv8c r-18u37iz r-1wtj0ep r-1ye8kvj r-1s2bzr4")\
                                        .find_all('div', class_="css-175oi2r r-18u37iz r-1h0z5md r-13awgt0")
                            comments = stats[0].text.strip() if len(stats) > 0 else "0"
                            repost = stats[1].text.strip() if len(stats) > 1 else "0"
                            likes = stats[2].text.strip() if len(stats) > 2 else "0"
                            views = stats[3].text.strip() if len(stats) > 3 else "0"
                        except:
                            comments = repost = likes = views = "0"
                    else:
                        try:
                            stats = p2[2].find('div', class_="css-175oi2r r-1kbdv8c r-18u37iz r-1wtj0ep r-1ye8kvj r-1s2bzr4")\
                                        .find_all('div', class_="css-175oi2r r-18u37iz r-1h0z5md r-13awgt0")
                            comments = stats[0].text.strip() if len(stats) > 0 else "0"
                            repost = stats[1].text.strip() if len(stats) > 1 else "0"
                            likes = stats[2].text.strip() if len(stats) > 2 else "0"
                            views = stats[3].text.strip() if len(stats) > 3 else "0"
                        except:
                            comments = repost = likes = views = "0"
    
                try:
                    post_link = "https://x.com" + p2[0].find_all('a')[2].get('href')
                except:
                    post_link = ""
    
                tweet_list.append([complainant_name, complainant_id, date_posted, written_post, post_link, have_img_vid, comments, repost, likes, views])
    
                if post_link and post_link not in all_tweets:
                    all_tweets.add(post_link)

                print(complainant_id)
    
                    # print({
                    #     "Name": complainant_name,
                    #     "Username": complainant_id,
                    #     "Date": date_posted,
                    #     "Tweet": written_post,
                    #     "Link": post_link,
                    #     "Media": have_img_vid,
                    #     "Comments": comments,
                    #     "Reposts": repost,
                    #     "Likes": likes,
                    #     "Views": views
                    # })
    
            except Exception as e:
                print("Tweet skipped due to structure mismatch:", e)
    
        # Check scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
        time.sleep(random.uniform(3, 6))
```
